```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder, Imputer
from xgboost import XGBClassifier
```

```python
df = pd.read_csv('dataset\horse-colic.data', delim_whitespace=True, header=None)
df.head(10).T
```

```python
df.info()
```

```python
df.isnull().any().sum()
```

```python
df.describe().T
```

```python
# sum of missing values (?) in each column
[(df[col].astype(str)=='?').sum() for col in df.columns]
```

```python
# To replace the missing value with NaN
df.replace(to_replace="?", value=np.nan, inplace=True)
df.head(10).T
```

```python
[col for col in df.columns if df[col].dtype == 'object']
```

```python
features = df.columns.to_list()[:-1]
label = df.columns.to_list()[-1]
X = df[features]
y = df[label]
X.shape, y.shape    
```

```python
y.value_counts().plot(kind='bar')
plt.show()
```

```python
def model(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state=7)
    xgb_model = XGBClassifier()
    print("\nModel", xgb_model)
    xgb_model.fit(X_train, y_train)
    y_pred = xgb_model.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    print("Accuracy :", score)
    
```

```python
# Encoding the label
labelEncoder = LabelEncoder()
labelEncoder.fit(y)
y_encoded = labelEncoder.transform(y)
```

```python
# To convert numpy array to pandas Series for creating bar plot
pd.Series(y_encoded).value_counts().plot(kind='bar')
# plt.show()
```

```python
# First experiment: Replacing all NaN values to 0 and
X1 = X.copy()
X1.replace(to_replace=np.nan, value=0, inplace = True)
X1 = X1.astype('float32')
X1.head(10).T
```

```python
model(X1, y_encoded)
```

```python
# Second Experiment:  Replacing all NaN values to 1 and
X2 = X.copy()
X2.replace(to_replace=np.nan, value=1, inplace = True)
X2 = X2.astype('float32')
X2.head(10).T
model(X2, y_encoded)
```

```python
# Third Experiment:  Keeping  all NaN values 
X3 = X.copy()
X3 = X3.astype('float32')
model(X3, y_encoded)
```

```python
# Fourth Experiment:  Replacing all NaN values with Imputer (default: mean)
X4 = X.copy()
imputer = Imputer()
X4 = imputer.fit_transform(X4)
X4 = X4.astype('float32')
model(X4, y_encoded)
```
